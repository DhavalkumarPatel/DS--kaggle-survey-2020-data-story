```python
#How does income vary based on the country of origin? Is the compensation same for different genders?

#Get the list of the countries (descending order of counts), selected first 30
a = list(c_index[0:30])

#Considering only Man and Woman gender for analysis
c_data = clean_data[(clean_data.Q2=="Man")|(clean_data.Q2=="Woman")&(clean_data.Q3.isin(a))].copy()

#Updating 'United States of America' to 'USA' to manage plots
c_data.loc[c_data['Q3']=="United States of America","Q3"] = "USA"
a[a.index("United States of America")] = "USA"

#Updating 'United Kingdom of Great Britain and Northern Ireland' to UK'
c_data.loc[c_data['Q3']=="United Kingdom of Great Britain and Northern Ireland","Q3"] = "UK"
a[a.index("United Kingdom of Great Britain and Northern Ireland")] = "UK"

#Determine the rank of countries based on the salaries of Man
c_pivot_data = clean_data[(clean_data.Q2=="Man")|(clean_data.Q2=="Woman")]
c_pivot = (pd.pivot_table(c_pivot_data, values='Q24', index=['Q3'], columns=['Q2'], aggfunc=np.mean)).sort_values(by="Man", ascending=False)
rank = list(c_pivot.index)
rank[rank.index("United States of America")] = "USA"
rank[rank.index("United Kingdom of Great Britain and Northern Ireland")] = "UK"

#Average salaries for man and woman in the dataset
S_man = clean_data[(clean_data.Q2=="Man")]['Q24'].mean()
S_woman = clean_data[(clean_data.Q2=="Woman")]['Q24'].mean()

#Initialize the matplotlib figure with a set size
fig = plt.subplots(figsize=(42,15))

#Plot 1
g_plot = sns.boxplot(x="Q3", y="Q24", hue="Q2", data = c_data, showfliers=False, order=a, saturation=1, palette='Set3', linewidth=3.5);
plt.axhline(y=S_man, color='blue', linestyle='--')
plt.text(27.3,S_man+5000,'Man (AvgSalary)', fontsize=18, weight="bold")
plt.axhline(y=S_woman, color='red', linestyle='-.')
plt.text(26.9,S_woman+5000,'Woman (AvgSalary)', fontsize=18, weight="bold")

#Labeling
new_labels = [item.get_text() + "_({})".format(rank.index(item.get_text())+1) for item in g_plot.get_xticklabels()];
g_plot.set_xticklabels(new_labels, rotation=90, fontsize=20);
plt.xlabel('Top 30 represented COUNTRIES with Salary Rank in (brackets)', fontsize=25);
plt.ylim(0,300000)
plt.tick_params(axis="y", labelsize=15)
plt.ylabel('Yearly Compensation (US $0-300K)', fontsize=25);
plt.title('Salary Comparison Plot based on "Country" and "Gender"', fontsize=30);
plt.legend(loc=1, prop={'size': 30});
```

**<font color='orange'>2.2 Two sample t-test for comparing average salaries</font>**

**What would we like to know?**
We would like to compare the salaries of male and female gender and check if it is significantly different. We have different variances among these two groups, So we will conduct unequal variances t-test.

**Our Parameter:** Mean Salary.

**Hypothesis:** 

1. **H<sub>0</sub>**: The difference in the average salaries is zero.
>> mean(male_salaries)-mean(female_salaries)=0

2. **H<sub>a</sub>**: The alternative hypothesis is that the difference is not zero.
>> mean(male_salaris)-mean(female_salaries)!=0

As we highlighted above, from the box plot we can't tell that the difference is statistically significant as the plot presents the spread in the same scale as that of the data. We need to include the variances of the mean for a better comparison. 

**Check Assumptions and Conditions for t-test:**
> <font color="blue">**Randomization**: The data can be considered a random sample selected from its true population.

> <font color="blue">**Independence**: The survey was independent. The data collected will have no dependence on the answers from other participants.

> <font color="red">**Normal Distribution Condition**: The code below presents two histograms for understanding the distribution of male and female salaries. We clearly see that the distributions are not normal or close to normal. 

**We cannot perform two sample t-test as all the required conditions were not cleared.**








```python
#Plot for Male Salary Distribution
fig = plt.subplots(figsize=(10,5))
sal_data.loc[(sal_data["Q2"]=="Man"),"Q24"].hist()
sal_data.loc[(sal_data["Q2"]=="Woman"),"Q24"].hist()
plt.legend(["Man","Woman"]);
```

**<font color='orange'>2.3 Bootstrap for Comparing Means</font>**

What we observed in the previous section was that the distributions were not normal. In order to develop a normal distribution, we will use CLT. The conditions for applying CLT are met as we know that the data-points are independent and a larger sample size (>100) will be selected because of the skewness in the data.

For bootstapping, the main parameter to consider for our study is sample size. For the presented results, 100% of the population size was used as sample size. Although, several sample sizes ranging from 5% to 90% were studied for comparison.

As the sample size increases, the distribution plot gets narrower. The bootstrapped distributions affect the distribution of difference of means as well as our t-test results. But, for this dataset, we get similar conclusion that there is a significant difference in salaries of man and woman. 

The code below focuses on two requirements:

1. The bootstrap distribution plots of male and female members. The output also specifies the mean and standard error of bootstrapped data and the original data.

2. The distribution plot for the difference of means. (I have also presented the expected normal pdf of difference of means for visualization).

**The output might differ from every run because of the random selection of data-points while bootstrapping.**



```python
#Separated Male and Female Salaries
sal_data_m = sal_data[(sal_data.Q2=="Man")]["Q24"]
sal_data_w = sal_data[(sal_data.Q2=="Woman")]["Q24"]

#Set number of iterations
num_iterations = 1000

#Set sample size 
num_samples_m = int(1*len(sal_data_m)) #100% sampling
num_samples_w = int(1*len(sal_data_w)) #100% sampling
print("Male Sample size: {}, Female sample size: {}\n".format(num_samples_m, num_samples_w))

#Bootstrap for the male and female salaries
stat_m = []
stat_w = []
diff = []
for i in range(num_iterations):
  boot_man = resample(sal_data_m, n_samples=num_samples_m)
  stat_m.append(boot_man.mean())

  boot_woman = resample(sal_data_w, n_samples=num_samples_w)
  stat_w.append(boot_woman.mean())

#Display the mean and standard deviation of bootstrapped and original data-points 
print("Mean salary for man- bootstrapped: {}, sample :{}".format(mean(stat_m), mean_m))
print("Standard Deviation man- bootstrapped: {}, sample mean/sqrt(samples): {}\n".format(stdev(stat_m),std_m/math.sqrt(num_samples_m)))

print("Mean salary for woman- bootstrapped: {},sample:{}".format(mean(stat_w), mean_w))
print("Standard Deviation woman- bootstrapped: {}, sample standard deviation/sqrt(samples): {}\n".format(stdev(stat_w),std_w/math.sqrt(num_samples_w)))

#Plot the bootstrapped distribution alongside each other
fig = plt.subplots(figsize=(10,5))
plt.hist(stat_m, color="blue");
plt.hist(stat_w, color="orange");
plt.xlabel("Salary Mean from Bootstraped Samples")
plt.legend(["Man","Woman"]);
plt.title("Bootstrapped Distributions");
plt.show()

#Reorder 1000 times the bootstrapped mean salaries and compute their difference.
for i in range(1000):
  m_means = resample(stat_m, n_samples=1000)
  w_means = resample(stat_w, n_samples=1000)
  di = mean(m_means) - mean(w_means)
  diff.append(di)

#Theoretical average and standard error of difference of means 
mean_diff = mean(stat_m)-mean(stat_w)
std_diff = math.sqrt((stdev(stat_m)**2)/(1000) + (stdev(stat_w)**2)/(1000))

print("\nTheoretical mean of difference of means, [(Mean of bootstrapped Male Salary) - (Mean of bootstrapped Female Salary)]: {}".format(mean_diff))
print("Actual mean of difference of means of bootstrapped samples from the above distributions: {}".format(mean(diff)))

print("\nTheoretical SE of difference of means, [Formula used is-> sqrt((sm^2/nm) + (sf^2/nf))]: {}".format(std_diff))
print("Actual std of difference of means of bootstrapped samples from the above distributions: {}". format(stdev(diff)))

#Compute and print the 95% confidence interval for the range of mean difference of salaries
print("\nWe can interpret with 95% confidence that on average the difference of salaries between male and female is in between: [{:.3f},{:.3f}]\n\n".format((mean_diff - 2.263*std_diff), (mean_diff + 2.263*std_diff)))

#Plot the Distribution of difference of means alongwith expect normal pdf
fig2 = plt.subplots(figsize=(8,5))
plt.hist(diff, density=True, color="green");
plt.xlabel("Difference of means from Bootstraped Samples")
plt.title("Difference of Means Distributions Plot with normal theoretical pdf");

x_min, x_max = plt.xlim()
x_data = np.linspace(x_min, x_max, 100)
p = norm.pdf(x_data, mean_diff, std_diff)
plt.plot(x_data, p, 'black', linewidth=2.5);
plt.legend(["Expected Normal PDF","Histogram from bootstrap"]);
```
